```
from pyspark.sql.functions import *
from pyspark.sql.types import *

# Define a path to the data source files in a python variable.
# Create a Spark Dataframe.
#fileName = "2015_2_clickstream.json"
try:
  spark.conf.get("advancingdlt.pipeline.fileName")
except:
  spark.conf.set("advancingdlt.pipeline.fileName", "2015_2_clickstream.json")
  
fileName = spark.conf.get("advancingdlt.pipeline.fileName")
jsonPath = f"/databricks-datasets/wikipedia-datasets/data-001/clickstream/raw-uncompressed-json/{fileName}"
options = {'inferSchema': True}
tableName = 'ClickStream'
df = spark.read.options(**options).json(jsonPath)
display(df)
```
